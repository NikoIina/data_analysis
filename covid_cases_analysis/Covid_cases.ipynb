```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

worldwide = pd.read_excel('covid_worldwide.xlsx')

cases_filter = (worldwide['country'] == 'Germany') & (worldwide['indicator'] == 'cases')
all_cases = worldwide.loc[cases_filter, ['weekly_count', 'year_week']]

plt.plot(all_cases['year_week'],all_cases['weekly_count'], label = 'Количество случаев', color='blue', alpha=0.4)
av_rolling = all_cases.rolling(window=10).mean()
plt.plot(all_cases['year_week'], av_rolling, label = 'Среднее скользящее', color='red')

plt.xlabel('Дата')
plt.ylabel('Выявленные случаи заболевания')
plt.title('Случаи заражения COVID-19 в Германии')
plt.xticks(np.arange(0,all_cases['year_week'].count()+1, 7))
plt.legend(loc='upper left')
plt.gcf().autofmt_xdate()

plt.show()

plt.hist(all_cases['weekly_count'])

plt.xlabel('Выявленные случаи заболевания')
plt.title('График распределения величины')

plt.show()
```
